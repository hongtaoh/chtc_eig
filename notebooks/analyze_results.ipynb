```python
import pandas as pd 
import numpy as np 
import json 
from pathlib import Path
```

```python
results_dir = Path("../results")

all_results = []

for file in results_dir.glob("*.json"):
    with file.open("r", encoding="utf-8") as f:
        data = json.load(f)
        all_results.extend(data)   # assumes each file is a list
```

```python
df = pd.DataFrame(all_results)
df.head()
```

```python
summary = (
    df.groupby("condition")
    .agg(avg_num_try=("num_try", "mean"), correct_rate=("correct", "mean")))

summary
```
